```python
# =========================================================
# AUGMENT TRAIN / VAL / TEST SEPARATELY (NO DATA LEAKAGE)
# =========================================================

import os
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img

# =========================
# PATHS
# =========================

train_dir = "/content/drive/MyDrive/Chickpeas Classification/Dataset_split/train"
val_dir   = "/content/drive/MyDrive/Chickpeas Classification/Dataset_split/val"
test_dir  = "/content/drive/MyDrive/Chickpeas Classification/Dataset_split/test"

# Augmented output folders
aug_train_dir = "/content/drive/MyDrive/Chickpeas Classification/Augmented/train"
aug_val_dir   = "/content/drive/MyDrive/Chickpeas Classification/Augmented/val"
aug_test_dir  = "/content/drive/MyDrive/Chickpeas Classification/Augmented/test"

# =========================
# TARGET IMAGES
# =========================

TRAIN_TARGET_PER_CLASS = 3000
VAL_TARGET_TOTAL = 750
TEST_TARGET_TOTAL = 750

# =========================
# AUGMENTATION SETTINGS
# =========================

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    shear_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# =========================================================
# FUNCTION
# =========================================================

def augment_dataset(input_dir, output_dir, target_per_class=None, target_total=None):

    os.makedirs(output_dir, exist_ok=True)

    class_names = os.listdir(input_dir)

    # =====================================================
    # TRAIN STYLE (TARGET PER CLASS)
    # =====================================================
    if target_per_class is not None:

        for class_name in class_names:

            class_input = os.path.join(input_dir, class_name)
            class_output = os.path.join(output_dir, class_name)

            os.makedirs(class_output, exist_ok=True)

            images = os.listdir(class_input)

            # Copy original images first
            for img_name in images:
                src = os.path.join(class_input, img_name)
                dst = os.path.join(class_output, img_name)

                if not os.path.exists(dst):
                    img = load_img(src)
                    img.save(dst)

            current_count = len(os.listdir(class_output))

            print(f"\nClass: {class_name}")
            print(f"Original Images: {len(images)}")

            while current_count < target_per_class:

                img_name = random.choice(images)

                img_path = os.path.join(class_input, img_name)

                img = load_img(img_path)
                x = img_to_array(img)
                x = x.reshape((1,) + x.shape)

                aug_iter = datagen.flow(
                    x,
                    batch_size=1
                )

                aug_img = next(aug_iter)[0].astype('uint8')

                save_name = f"aug_{current_count}.jpg"

                array_to_img(aug_img).save(
                    os.path.join(class_output, save_name)
                )

                current_count += 1

            print(f"Final Images: {current_count}")

    # =====================================================
    # VAL / TEST STYLE (TARGET TOTAL)
    # =====================================================
    elif target_total is not None:

        all_images = []

        for class_name in class_names:

            class_input = os.path.join(input_dir, class_name)
            class_output = os.path.join(output_dir, class_name)

            os.makedirs(class_output, exist_ok=True)

            images = os.listdir(class_input)

            for img_name in images:

                src = os.path.join(class_input, img_name)
                dst = os.path.join(class_output, img_name)

                if not os.path.exists(dst):
                    img = load_img(src)
                    img.save(dst)

                all_images.append((class_name, img_name))

        current_total = sum([
            len(os.listdir(os.path.join(output_dir, c)))
            for c in class_names
        ])

        print(f"\nOriginal Total: {current_total}")

        while current_total < target_total:

            class_name, img_name = random.choice(all_images)

            img_path = os.path.join(input_dir, class_name, img_name)

            img = load_img(img_path)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)

            aug_iter = datagen.flow(
                x,
                batch_size=1
            )

            aug_img = next(aug_iter)[0].astype('uint8')

            save_name = f"aug_{current_total}.jpg"

            array_to_img(aug_img).save(
                os.path.join(output_dir, class_name, save_name)
            )

            current_total += 1

        print(f"Final Total: {current_total}")

# =========================================================
# RUN
# =========================================================

# TRAIN → 3000 per class
augment_dataset(
    train_dir,
    aug_train_dir,
    target_per_class=TRAIN_TARGET_PER_CLASS
)

# VAL → total 750
augment_dataset(
    val_dir,
    aug_val_dir,
    target_total=VAL_TARGET_TOTAL
)

# TEST → total 750
augment_dataset(
    test_dir,
    aug_test_dir,
    target_total=TEST_TARGET_TOTAL
)

print("\nALL AUGMENTATION COMPLETED SUCCESSFULLY")
```

```python
# ==========================================
# CHECK IMAGE COUNT IN TRAIN / VAL / TEST
# ==========================================

import os

# Augmented folders
aug_train_dir = "/content/drive/MyDrive/Chickpeas Classification/Augmented/train"
aug_val_dir   = "/content/drive/MyDrive/Chickpeas Classification/Augmented/val"
aug_test_dir  = "/content/drive/MyDrive/Chickpeas Classification/Augmented/test"

# ==========================================
# FUNCTION
# ==========================================

def count_images(folder_path, dataset_name):

    print(f"\n===== {dataset_name} =====")

    total = 0

    class_names = os.listdir(folder_path)

    for class_name in class_names:

        class_path = os.path.join(folder_path, class_name)

        count = len(os.listdir(class_path))

        print(f"{class_name}: {count} images")

        total += count

    print(f"TOTAL: {total} images")


# ==========================================
# CHECK COUNTS
# ==========================================

count_images(aug_train_dir, "TRAIN DATASET")
count_images(aug_val_dir, "VALIDATION DATASET")
count_images(aug_test_dir, "TEST DATASET")
```
